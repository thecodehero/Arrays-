```python
def timesOccursIn(k,A):
    if len(A) == 0: return 0; #check if array is empty. Stops recursive calls when reaches end of array A.
    else:
        if A[0] == k: #if first element equals to k
            return 1 + timesOccursIn(k,A[1:]) #add 1 to the amount of k in array. Counts number of k.
        else:
            return timesOccursIn(k,A[1:]) 
        
        
        
        
    
print(timesOccursIn(5,[1,2,5,3,6,5,3,5,5,4]))
print(timesOccursIn(5,[]))
print(timesOccursIn(5,[5,4,2,5,5]))
```

```python
import time

def selectionSort(A):
     for i in range(len(A)):
            imin = findMin(i,A)
            swap(i,imin,A)
            
def findMin(i, A):
    imin = i
    for j in range(i+1,len(A)):
        if A[j] < A[imin]:
            imin = j
    return imin

def swap(i, j, A):
    tmp = A[i]
    A[i] = A[j]
    A[j] = tmp
    
def insertionSort(A):
    for i in range(1,len(A)):
        insert(A[i],A,i)

def insert(v, A, hi):
    for i in range(hi-1,-1,-1):
        if v >= A[i]:
            A[i+1] = v
            return
        A[i+1] = A[i]
    A[0] = v
    
def mergeSort(A):
    if len(A) <= 1:
        return
    mid = len(A)//2
    half1 = A[:mid]
    half2 = A[mid:]
    mergeSort(half1)
    mergeSort(half2)
    merge(half1,half2,A)

def merge(h1, h2, A):
    i=0; j1=0; j2=0
    while j1<len(h1) and j2<len(h2):
        if h1[j1] < h2[j2]:
            A[i] = h1[j1]
            j1 += 1; i += 1
        else:
            A[i] = h2[j2]
            j2 += 1; i += 1 
    while j1 < len(h1):
        A[i] = h1[j1]
        j1 += 1; i += 1
    while j2 < len(h2):
        A[i] = h2[j2]
        j2 += 1; i += 1
        
def quickSort(A):
    quickSortRec(A,0,len(A))

def quickSortRec(A, lo, hi):
    # sorts A[lo:hi]
    if hi-lo <= 1:
        return
    iPivot = partition(A,lo,hi)
    quickSortRec(A,lo,iPivot)
    quickSortRec(A,iPivot+1,hi)
    
def partition(A, lo, hi):
    pivot = A[lo]
    B = [0 for i in range(lo,hi)]
    loB = 0
    hiB = len(B)-1
    for i in range(lo+1,hi):
        if A[i] < pivot:
            B[loB] = A[i]
            loB += 1
        else:
            B[hiB] = A[i]
            hiB -= 1
    B[loB] = pivot
    for i in range(len(B)):
        A[lo+i] = B[i]
    return lo+loB

def randomArr(n):
    array = []
    for i in range(n):
        array.append(random.randint(0,9))
    return array

def timeCalc(f):
    t = time.time()
    f
    t = time.time() - t
    return t

def meanTime(t1,t2,t3):
    mean = (t1+t2+t3/3)
    return mean

A100 = randomArr(100)
B100 = randomArr(100)
C100 = randomArr(100)

A1000 = randomArr(1000)
B1000 = randomArr(1000)
C1000 = randomArr(1000)

A10000 = randomArr(10000)
B10000 = randomArr(10000)
C10000 = randomArr(10000)

A100000 = randomArr(100000)
B100000 = randomArr(100000)
C100000 = randomArr(100000)

#selection sort
s1 = meanTime(timeCalc(selectionSort(A100)),timeCalc(selectionSort(B100)),timeCalc(selectionSort(C100)))
s2 = meanTime(timeCalc(selectionSort(A1000)),timeCalc(selectionSort(B1000)),timeCalc(selectionSort(C1000)))
s3 = meanTime(timeCalc(selectionSort(A10000)),timeCalc(selectionSort(B10000)),timeCalc(selectionSort(C10000)))
s4 = meanTime(timeCalc(selectionSort(A100000)),timeCalc(selectionSort(B100000)),timeCalc(selectionSort(C100000)))
print(s1)
print(s2)
print(s3)
print(s4)

#insertion sort
i1 = meanTime(timeCalc(insertionSort(A100)),timeCalc(insertionSort(B100)),timeCalc(insertionSort(C100)))
i2 = meanTime(timeCalc(insertionSort(A1000)),timeCalc(insertionSort(B1000)),timeCalc(insertionSort(C1000)))
i3 = meanTime(timeCalc(insertionSort(A10000)),timeCalc(insertionSort(B10000)),timeCalc(insertionSort(C10000)))
i4 = meanTime(timeCalc(insertionSort(A100000)),timeCalc(insertionSort(B100000)),timeCalc(insertionSort(C100000)))
print(i1)
print(i2)
print(i3)
print(i4)

#merge sort
m1 = meanTime(timeCalc(mergeSort(A100)),timeCalc(mergeSort(B100)),timeCalc(mergeSort(C100)))
m2 = meanTime(timeCalc(mergeSort(A100)),timeCalc(mergeSort(B100)),timeCalc(mergeSort(C100)))
m3 = meanTime(timeCalc(mergeSort(A100)),timeCalc(mergeSort(B100)),timeCalc(mergeSort(C100)))
m4 = meanTime(timeCalc(mergeSort(A100)),timeCalc(mergeSort(B100)),timeCalc(mergeSort(C100)))
print(m1)
print(m2)
print(m3)
print(m4)

#quick sort
q1 = meanTime(timeCalc(quickSort(A100)),timeCalc(quickSort(B100)),timeCalc(quickSort(C100)))
q2 = meanTime(timeCalc(quickSort(A1000)),timeCalc(quickSort(B1000)),timeCalc(quickSort(C1000)))
q3 = meanTime(timeCalc(quickSort(A10000)),timeCalc(quickSort(B10000)),timeCalc(quickSort(C10000)))
q4 = meanTime(timeCalc(quickSort(A100000)),timeCalc(quickSort(B100000)),timeCalc(quickSort(C100000)))
print(q1)
print(q2)
print(q3)
print(q4)
```
